# converting Timing columns to datetime object 
> steps :
1. Converting `Available From` and `Available Till` columns to datetime format.
2. Creating new column `Number of Hrs. Available` to show total number of hours of availability by subtracting time delta of `Available Till`  from time delta of `Available From`
3. Removing days part from the `Number of Hrs. Available` column

```python
df.columns
```

```python
# Convert the time column to datetime data type
df['Available From'] = pd.to_datetime(df['Available From'], format='%I:%M %p').dt.time
df['Available Till'] = pd.to_datetime(df['Available Till'], format='%I:%M %p').dt.time
```

```python
df.head()
```

```python
df['Number of Hrs. Available'] = (pd.to_timedelta(df['Available Till'].astype(str)) - pd.to_timedelta(df['Available From'].astype(str)))
```

```python
df
```

```python
df.info()
```

```python
hrs = []
for i in df['Number of Hrs. Available'] :
    b = re.findall(r'days (.+):00$',str(i))
    if len(b) > 0 :
        if b[0][0] == '+' :
            hrs.append(b[0][1:])
        else :
            hrs.append(b[0])
    else :
        hrs.append(i)
    
```

```python
df['Number of Hrs. Available'] = hrs
```

```python
df.rename({'Number of Hrs. Available' : 'Number of Hrs. Available (HH:MM)'},inplace = True)
```
