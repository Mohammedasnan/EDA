```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

```python
dr_names = []
spl_col = []
experience = []
Education = []
clinic_fee = []
online_fee = []
clinic_name = []
clinic_location = []
recent_consult = []
start_time = []
end_time = []
city = []
```

```python
specialization = ['Dentist', 'Dermatologist', 'Dietitian/Nutritionist', 'General+Physician', 'Orthopedist', 'Gynaecologist', 'Homeopathy', 'Psychologist', 'Ayurveda']
```

```python
request_header = {'Content-Type': 'text/html; charset=UTF-8', 
                  'User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0', 
                  'Accept-Encoding': 'gzip, deflate, br'}

url = 'https://www.lybrate.com/search?find=Dentist&cityName=Hyderabad&seoRequest=true&page=1'
# running loops for the multiple pages 
for j in specialization:
    count = 0
    for i in range(1,6):
        # connection website
        page = requests.get('https://www.lybrate.com/search?find='+j+'&cityName=Hyderabad&seoRequest=true&page='+str(i))

        # converting str data to html format

        soup = BeautifulSoup(page.text)
        # dr names
        scrape_names = soup.find_all('h2',class_ = 'doctorCard_doctorName__independent__VjX6P')
        for i in scrape_names :
            dr_names.append(i.text)
            spl_col.append(j)
            city.append('Hyderabad')
        
        # experience 
        for i in soup.find_all('div',class_="doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua"):
            
            a = i.text
            if re.findall('(\d+) Years Exp.',a) :
                
                experience.append(re.findall('(\d+) Years Exp.',a)[0])
            else :
                experience.append(np.nan)
        
                            
        # clinic fee 
        for i in soup.find_all('div',class_="doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua") :
    
            if re.findall(r'(₹\s\d+)\sat\sclinic',i.text) :
            
                clinic_fee.append(re.findall(r'(₹\s\d+)\sat\sclinic',i.text)[0])
            else :
                clinic_fee.append(np.nan)
            
            b = re.findall(r'(₹\s\d+)\sonline',i.text)
            if len(b) > 0 :
                online_fee.append(b[0])
            else :
                online_fee.append(np.nan)
        
                            
        # Education
        for i in soup.find_all('div',class_ = 'doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua'):
            if re.findall('Exp\.([A-Za-z\,\s\.\-\(\)\&]+)\₹',i.text) :
                Education.append(re.findall('Exp\.([A-Za-z\,\s\.\-\(\)\&]+)\₹',i.text)[0])
            else :
                Education.append(np.nan)
                       
        #  clinic 
                            
        
        for i in soup.find_all('div',class_="doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua") :
            a = i.text
            if re.findall('\₹\s\d+\sonline(.+derabad)',a) :
                
                loc = re.findall('\₹\s\d+\sonline(.+derabad)',a)[0]
                clinic_name.append(loc.split(',')[0].strip())
                clinic_location.append(loc.split(',')[1].strip())
            else: 
                    if re.findall('\₹\s\d+\sat\sclinic(.+derabad)',a) :

                        loc = re.findall('\₹\s\d+\sat\sclinic(.+derabad)',a)[0]
                        clinic_name.append(loc.split(',')[0].strip())
                        clinic_location.append(loc.split(',')[1].strip())
                    
                    else :
                        clinic_name.append(np.nan)
                        clinic_location.append(np.nan)
                        
        
        
        # recent
        
        for i in soup.find_all('div',class_="doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua"):
            a = i.text
            b = re.findall(r'Recently consulted for:(.+)Avai|Recently consulted for:(.+)See',a)
            if len(b) > 0 :
                if b[0][0] == '' :
                    recent_consult.append(b[0][1])
                else :
                    recent_consult.append(b[0][0])
            else :
                recent_consult.append(np.nan)
                      
        # timings
                            
        
        for i in soup.find_all('div',class_="doctorCard_cardDesc__bq9bD doctorCard_cardOverflow__4D4Ua"):
            a = i.text
            
            b = re.findall(r'\d\d:\d\d\sAM|\d\d:\d\d\sPM',a)
            if len(b) > 0 :
                start_time.append(b[0])
                end_time.append(b[1])
            else :
                start_time.append(np.nan)
                end_time.append(np.nan)
        count += 1
        print(f'{j} Page {count}')
        print(f'Dr names {len(dr_names)}')
        print(f'Spl {len(spl_col)}')
        print(f'Online_fee {len(online_fee)}')
        print(f'clinic_fee {len(clinic_fee)}') 
        print(f'Experience {len(experience)}')
        print(f'Education {len(Education)}')
        print(f'Clinic Name {len(clinic_name)}')
        print(f'Clinic Location {len(clinic_location)}')
        print(f'recent_consult {len(recent_consult)}')
        print(f'start_time {len(start_time)}')
        print(f'end_time {len(end_time)}')
        print('\n')
    
    
```

```python
d = {'Name':dr_names,
     'Specialization' : spl_col,
    'Experience (in years)':experience,
    'Education':Education,
    'Fee at clinic':clinic_fee,
    'Online Fee':online_fee,
     'Clinic Name':clinic_name,
     'Area':clinic_location,
     'Recently Consulted for':recent_consult,
     'Available From':start_time,
     'Available Till' : end_time,
     'City' : city
    }
df = pd.DataFrame(d)
df
```
